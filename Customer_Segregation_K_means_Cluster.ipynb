```python
# prompt: plot graphs for age groups and annual income we only have age column which has individual ages. use age groups and average annual income in histogram

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/Mall_Customers.csv')
df.head()

print(df.columns.values)

# Data Preprocessing

df.drop(columns=['CustomerID'], inplace=True)
df.head()


df['Genre'].replace(['Male', 'Female'], [1, 0], inplace=True)
df.head()

# Create age groups
df['Age_Group'] = pd.cut(df['Age'], bins=[18, 25, 35, 45, 55, 65], labels=['18-25', '25-35', '35-45', '45-55', '55-65'])

# Calculate average annual income for each age group
df_grouped = df.groupby('Age_Group')['Annual Income (k$)'].mean().reset_index()

# Plot histogram
plt.bar(df_grouped['Age_Group'], df_grouped['Annual Income (k$)'])
plt.xlabel('Age Group')
plt.ylabel('Average Annual Income (k$)')
plt.title('Average Annual Income by Age Group')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt
import numpy as np

# Assuming df_scaled is the standardized DataFrame
df_scaled = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()



```

Ideal number of clusters = 5 according to Elbow method

```python
# Silhouette Score
silhouette_scores = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    cluster_labels = kmeans.fit_predict(df_scaled)
    silhouette_avg = silhouette_score(df_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis')
plt.show()
```

Ideal number of clusters = 5 According to Silhouette Score

```python

# Davies-Bouldin Index
db_scores = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    cluster_labels = kmeans.fit_predict(df_scaled)
    db_score = davies_bouldin_score(df_scaled, cluster_labels)
    db_scores.append(db_score)

plt.plot(range(2, 11), db_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Davies-Bouldin Index')
plt.title('Davies-Bouldin Analysis')
plt.show()
```

Ideal number of clusters = 5 According to David-Bouldin Analysis

```python
# prompt: add centroids to the same code written above, write the kmeans code again nothing before that

# Choose the number of clusters
num_clusters = 5

# Create a KMeans instance
kmeans = KMeans(n_clusters=num_clusters, random_state=0)

# Fit the KMeans model to the data
df_scaled = df[['Annual Income (k$)', 'Spending Score (1-100)']]
kmeans.fit(df_scaled)

# Get cluster labels for each data point
labels = kmeans.labels_

# Add cluster labels to the DataFrame
df['Cluster'] = labels

# Print the cluster centers
print(kmeans.cluster_centers_)

# Visualize the clusters
plt.scatter(df_scaled['Annual Income (k$)'], df_scaled['Spending Score (1-100)'], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', label='Centroids')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('K-Means Clustering with {} Clusters'.format(num_clusters))
plt.legend()
plt.show()
```

```python
import numpy as np

# Calculate the average age for each cluster
avg_age = df.groupby('Cluster')['Age'].mean()

# Choose the number of clusters
num_clusters = 5

# Create a KMeans instance
kmeans = KMeans(n_clusters=num_clusters, random_state=0)

# Fit the KMeans model to the data
df_scaled = df[['Annual Income (k$)', 'Spending Score (1-100)']]
kmeans.fit(df_scaled)

# Get cluster labels for each data point
labels = kmeans.labels_

# Add cluster labels to the DataFrame
df['Cluster'] = labels

# Print the cluster centers
print(kmeans.cluster_centers_)

# Visualize the clusters along with average age
plt.figure(figsize=(10, 8))

# Scatter plot of data points
plt.scatter(df_scaled['Annual Income (k$)'], df_scaled['Spending Score (1-100)'], c=labels, cmap='viridis')

# Plot centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', label='Centroids')

# Plot average age for each cluster
for i in range(num_clusters):
    plt.text(kmeans.cluster_centers_[i, 0], kmeans.cluster_centers_[i, 1], f'Avg Age: {avg_age[i]:.1f}', color='black', fontsize=10)

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('K-Means Clustering with {} Clusters'.format(num_clusters))

# Remove x-axis and y-axis numbering
plt.xticks([])
plt.yticks([])

plt.legend()
plt.show()
```
